```python
import requests
import os

# Base URL common to all files
url_common = "https://data.nber.org/cps-basic3/dta/"

# Function to download a file
def download_file(url, save_path):
    try:
        response = requests.get(url)
        response.raise_for_status()  # Check for HTTP request errors
        with open(save_path, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {save_path}")
    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error occurred: {http_err}")
    except Exception as err:
        print(f"Other error occurred: {err}")

save_folder = "/Users/tranguyen/Library/CloudStorage/OneDrive-YaleUniversity/online job ads_china/data/CPS"

# Define the range of years and months you want to download
years = range(1992, 2013)  # Example range of years
months = range(1, 13)  # January to December

# Iterate over the years and months to generate URLs and download files
for y in years:
    # Create a subfolder for each year
    year_folder = os.path.join(save_folder, str(y))
    os.makedirs(year_folder, exist_ok=True)  # Create year folder if it doesn't exist
    
    for m in months:
        # Format the month with leading zero if necessary
        month_str = str(m).zfill(2)
        # Create the file URL
        url = url_common + str(y) + "/" + "cpsb" + str(y) + month_str + ".dta"
        # Define the file name and path to save
        file_name = f"cpsb{y}{month_str}.dta"
        save_path = os.path.join(year_folder, file_name)
        # Download the file
        download_file(url, save_path)
```
